```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")  # load a pretrained model (recommended for training)
```

# Train Model

```python
results = model.train(data='/content/ultralytics/ultralytics/cfg/datasets/pothole.yaml', imgsz=640, epochs=100, batch=32, name="custom")
```

```python
results = model.predict(source="/content/test_pothole_video.mp4", save=True)
```
